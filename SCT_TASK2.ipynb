```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("train.csv")

# Show top rows
df.head()
```

```python
# Drop columns with too many missing values
df.drop(columns=['Cabin'], inplace=True)

# Fill missing Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Drop rows with missing Embarked
df.dropna(subset=['Embarked'], inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')

df = df.dropna(subset=['survived'])

custom_colors = ['#FF6F61', '#2EC4B6']

plt.figure(figsize=(8, 6))
ax = sns.countplot(x='survived', data=df, palette=custom_colors)

ax.set_xticklabels(['Did Not Survive', 'Survived'], fontsize=12)

plt.title('🛳️ Titanic Survival Count – A Vibrant View', fontsize=16, fontweight='bold', color='#333')
plt.xlabel('Survival Status', fontsize=13, fontweight='semibold')
plt.ylabel('Passenger Count', fontsize=13, fontweight='semibold')

for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='#333')

sns.despine()
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


df = sns.load_dataset('titanic').dropna(subset=['survived', 'sex'])


sns.set_style("whitegrid")


custom_palette = {'female': '#FF6F61', 'male': '#6B5B95'}

plt.figure(figsize=(10, 6))

ax = sns.countplot(
    x='survived',
    hue='sex',
    data=df,
    palette=custom_palette,
    edgecolor='black'
)

ax.set_xticklabels(['Did Not Survive', 'Survived'], fontsize=12)

plt.title('Titanic Survival by Gender', fontsize=18, fontweight='bold', color='darkgreen')
plt.xlabel('Survival Status', fontsize=14, fontweight='semibold')
plt.ylabel('Passenger Count', fontsize=14, fontweight='semibold')

for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}',
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom',
                    fontsize=11, fontweight='bold', color='black')

plt.legend(
    title='Gender',
    title_fontsize='13',
    fontsize='12',
    loc='upper right',
    frameon=True,
    edgecolor='black'
)

sns.despine()
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure data is clean
df = sns.load_dataset('titanic')
df = df.dropna(subset=['pclass', 'sex', 'survived'])

# Set a clean and stylish seaborn theme
sns.set_theme(style="whitegrid")

# Create catplot
g = sns.catplot(
    data=df,
    x='pclass',
    hue='survived',
    col='sex',
    kind='count',
    palette={0: '#FF6F61', 1: '#2EC4B6'},
    height=5,
    aspect=1
)

# Add a bold and centered super title
g.fig.suptitle(' Titanic Survival by Class and Gender', fontsize=18, fontweight='bold', color='darkslategray')
g.fig.subplots_adjust(top=0.85)

# Customize axes and titles for each subplot
g.set_axis_labels("Passenger Class", "Passenger Count")
g.set_titles(col_template="Gender: {col_name}", size=14, weight='bold', color='navy')
g.set_xticklabels(['1st', '2nd', '3rd'])

# Add value labels to each bar
for ax in g.axes.flat:
    for bar in ax.patches:
        height = int(bar.get_height())
        if height > 0:
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                height + 1,
                f'{height}',
                ha='center',
                va='bottom',
                fontsize=10,
                fontweight='bold',
                color='black'
            )

# Adjust layout and style
sns.despine()
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Clean and prepare the data
df = sns.load_dataset('titanic')
df = df.dropna(subset=['pclass', 'survived'])

# Set a clean theme
sns.set_style("whitegrid")

# Custom vibrant palette for survival status
custom_palette = {0: '#FF6F61', 1: '#2EC4B6'}  # Coral = No, Teal = Yes

# Create the figure
plt.figure(figsize=(10, 6))

# Create the countplot
ax = sns.countplot(x='pclass', hue='survived', data=df, palette=custom_palette, edgecolor='black')

# Set axis labels and title
plt.title('Titanic Survival by Passenger Class', fontsize=18, fontweight='bold', color='darkslateblue')
plt.xlabel('Passenger Class', fontsize=14, fontweight='semibold')
plt.ylabel('Passenger Count', fontsize=14, fontweight='semibold')
ax.set_xticklabels(['1st Class', '2nd Class', '3rd Class'], fontsize=12)

# Custom legend
plt.legend(title='Survived', labels=['No', 'Yes'], fontsize=11, title_fontsize=12, loc='upper right')

# Add count labels above each bar
for bar in ax.patches:
    count = int(bar.get_height())
    if count > 0:
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            count + 2,
            str(count),
            ha='center',
            fontsize=11,
            fontweight='bold',
            color='black'
        )

# Final layout tweaks
sns.despine()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count survival values
surv = df['survived'].value_counts()
labels = ['Did Not Survive', 'Survived']
counts = surv.values

# Combine labels with counts and percentages
def make_autopct(values):
    def inner_autopct(pct):
        total = sum(values)
        count = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n({count})'
    return inner_autopct

# Vibrant and contrasting colors
colors = ['#FF6B6B', '#4ECDC4']  # Vivid red and teal

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    counts,
    labels=labels,
    colors=colors,
    autopct=make_autopct(counts),
    startangle=140,
    explode=(0.05, 0.05),  # Explode both slices slightly for emphasis
    textprops={'fontsize': 13, 'weight': 'bold', 'color': 'black'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)

# Add bold, styled title
plt.title('Titanic Survival Rate – Visual Breakdown', fontsize=18, fontweight='bold', color='darkslategray')

# Perfect spacing
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute correlation matrix (numerical only)
corr = df.corr(numeric_only=True)

# Set up a visually appealing theme
sns.set_theme(style="white")

# Create the figure
plt.figure(figsize=(12, 8))  # Larger size for better readability

# Create heatmap with enhanced style
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="YlGnBu",             # Soft green-blue gradient
    linewidths=0.7,
    linecolor='white',
    square=True,
    cbar_kws={'shrink': 0.8, 'aspect': 30, 'label': 'Correlation'},
    annot_kws={"size": 11, "weight": "bold", "color": "black"}
)

# Title with elegant styling
plt.title('Titanic Data – Correlation Heatmap', fontsize=18, fontweight='bold', color='#2F4F4F')

# Axis ticks styling
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)
# Improve layout spacing
plt.tight_layout()
plt.show()
```
